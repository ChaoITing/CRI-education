```python
features_target_Collin = features_target.loc[features_target.censusTract.str.contains('^48085', regex= True, na=False), :]
# features_target_Collin['county'] = 'Collin'
features_target_Dallas = features_target.loc[features_target.censusTract.str.contains('^48113', regex= True, na=False), :]
# features_target_Dallas['county'] = 'Dallas'
features_target_Denton = features_target.loc[features_target.censusTract.str.contains('^48121', regex= True, na=False), :]
# features_target_Denton['county'] = 'Denton'
features_target_Tarrant = features_target.loc[features_target.censusTract.str.contains('^48439', regex= True, na=False), :]
# features_target_Tarrant['county'] = 'Tarrant'
print(len(features_target_Collin))
print(len(features_target_Dallas))
print(len(features_target_Denton))
print(len(features_target_Tarrant))
```

```python
from missingpy import KNNImputer
import random
random.seed(0)

imputer = KNNImputer(n_neighbors = 5)
features_target_Collin_imputed = imputer.fit_transform(features_target_Collin)
features_target_Dallas_imputed = imputer.fit_transform(features_target_Dallas)
features_target_Denton_imputed = imputer.fit_transform(features_target_Denton)
features_target_Tarrant_imputed = imputer.fit_transform(features_target_Tarrant)

features_target_Collin_imputed = pd.DataFrame(features_target_Collin_imputed, columns = features_target_Collin.columns)
features_target_Dallas_imputed = pd.DataFrame(features_target_Dallas_imputed, columns = features_target_Dallas.columns)
features_target_Denton_imputed = pd.DataFrame(features_target_Denton_imputed, columns = features_target_Denton.columns)
features_target_Tarrant_imputed = pd.DataFrame(features_target_Tarrant_imputed, columns = features_target_Tarrant.columns)

features_target_imputed = features_target_Collin_imputed.append([features_target_Dallas_imputed, features_target_Denton_imputed, features_target_Tarrant_imputed], ignore_index = True)

features_target_imputed.info()
```

```python
features_target_imputed.head(10)
```

## standardize

```python
from sklearn.preprocessing import MinMaxScaler

censusTract = features_target_imputed['censusTract']
to_standardize = features_target_imputed.drop('censusTract', axis = 1)

minmaxS = MinMaxScaler()
features_target_imputed_scale = minmaxS.fit_transform(to_standardize)

features_target_imputed_scale = pd.DataFrame(features_target_imputed_scale, columns = to_standardize.columns)

features_target_imputed_scale = pd.concat([censusTract, features_target_imputed_scale], axis = 1)
features_target_imputed_scale.head(10)
```

## simple linear regression

```python
#reverse 'Poverty_prob_index', 'MHLTH', 'PHLTH'
```

```python
features_target_imputed_scale['R_Poverty_prob_index'] = features_target_imputed_scale['Poverty_prob_index']*(-1)
features_target_imputed_scale['R_MHLTH'] = features_target_imputed_scale['MHLTH']*(-1)
features_target_imputed_scale['R_PHLTH'] = features_target_imputed_scale['PHLTH']*(-1)
features_target_imputed_scale.drop(['Poverty_prob_index', 'MHLTH', 'PHLTH'], axis = 1, inplace = True)
```

```python
features_target_imputed_scale.columns
```

```python
a = features_target_imputed_scale.iloc[:, 1:12].columns
b = features_target_imputed_scale.iloc[:, 12:16].columns
```

```python
features_target_imputed_scale
```

```python
from sklearn.linear_model import LinearRegression

coeff = []

for X in a:
    for y in b:
        lr = LinearRegression()    
        lr.fit(features_target_imputed_scale[[X]], features_target_imputed_scale[y])        
        coeff.append(lr.coef_)
coeff
```

```python
Pearson_correl_matrix = np.reshape(coeff, (11, 4))
Pearson_correl_matrix
```

```python
Indicator_outcome_Correl = pd.DataFrame(Pearson_correl_matrix, index=a, columns=b)
Indicator_outcome_Correl
```

```python
Indicator_outcome_Correl['Avg']=Indicator_outcome_Correl[['med_hhinc2016', 'R_Poverty_prob_index', 'R_MHLTH', 'R_PHLTH']].mean(axis=1)
Indicator_outcome_Correl
```

```python
Indicator_outcome_Correl['rescale_coeff'] = Indicator_outcome_Correl['Avg']*(11/Indicator_outcome_Correl['Avg'].sum(axis = 0))
Indicator_outcome_Correl['wj'] = (Indicator_outcome_Correl['rescale_coeff']+1)/2
Indicator_outcome_Correl['weight'] = Indicator_outcome_Correl['wj']/Indicator_outcome_Correl['wj'].sum(axis = 0)
Indicator_outcome_Correl
```

## Calculate the score for each census tract

```python
weights = Indicator_outcome_Correl.drop(['med_hhinc2016', 'R_Poverty_prob_index', 'R_MHLTH', 'R_PHLTH', 'Avg', 'rescale_coeff', 'wj'], axis = 1)
weights
```

```python
features = features_target_imputed_scale.iloc[:, 0:12]
features
```

```python
censusTract = features.censusTract
score = features.iloc[:, 1:12]*weights.weight.T
score = pd.concat([censusTract, score], axis = 1)
score
```

```python
score['index'] = score.iloc[:, 1:12].sum(axis = 1)
score['score'] = (score['index']-score['index'].min())/(score['index'].max()-score['index'].min())*100
score
```

```python
score['censusTract'] = score['censusTract'].astype(str)
score['censusTract'] = score['censusTract'].str[:11]
score
```
